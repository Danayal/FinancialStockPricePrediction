```python


def classifying(x):
    if x > 1: 
        if x < 10:
            return "Class F"
        if x < 100:
            return "Class E"
        if x < 1000:
            return "Class D"
        if x < 10000:
            return "Class C"
        if x < 100000:
            return "Class B"
        if x > 1000000:
            return "Class A"
    
    else:
        if x > -10:
            return "Class W"
        if x > -100:
            return "Class X"
        if x > -1000:
            return "Class Y"
        if x < 1000:
            return "Class Z"
        
        
            

        
   
```

```python
df["classes"] = df['2015 PRICE VAR [%]'].apply(classifying)
df
```
